```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
weathers_df = pd.read_csv('weatherdata.csv')
```

```python
weathers_df.head()
```

```python
weathers_df.info(null_counts=True)
```

```python
weathers_df.loc[~weathers_df.precipType.isnull(),'precipType']=1.0
weathers_df.loc[weathers_df.precipType.isnull(),'precipType']=0.0
weathers_df.drop(['windGustTime', 'windGust','visibility'], axis=1,inplace=True)
```

```python
weathers_df.info(null_counts=True)
```

```python
def convert_localhour(df,at):
    df[at] = pd.to_datetime(df[at],unit='s').dt.tz_localize('UTC').dt.tz_convert('Asia/Ho_Chi_Minh').dt.hour
    
def convert_localday(df,at):
    df[at] = pd.to_datetime(df[at],unit='s').dt.tz_localize('UTC').dt.tz_convert('Asia/Ho_Chi_Minh')
```

```python
convert_localhour(weathers_df,'sunriseTime')
convert_localhour(weathers_df,'sunsetTime')
convert_localhour(weathers_df,'uvIndexTime')
convert_localday(weathers_df,'time')
convert_localhour(weathers_df,'temperatureHighTime')
convert_localhour(weathers_df,'temperatureLowTime')
convert_localhour(weathers_df,'temperatureMinTime')
convert_localhour(weathers_df,'temperatureMaxTime')
convert_localhour(weathers_df,'apparentTemperatureMinTime')
convert_localhour(weathers_df,'apparentTemperatureMaxTime')
```
